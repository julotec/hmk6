```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Wczytanie danych
data_2d = pd.read_csv('/drive/hmk6/data_2d.csv')
data_mnist = pd.read_csv('/drive/hmk6/mnist.csv')

# Wizualizacja danych 2D
plt.scatter(data_2d.iloc[:, 1], data_2d.iloc[:, 2])
plt.title('2D Data')
plt.xlabel('Column 1')
plt.ylabel('Column 2')
plt.show()

# Wizualizacja danych MNIST (nie jest to najlepsza wizualizacja dla danych MNIST, ale można użyć dla celów demonstracyjnych)
image = data_mnist.iloc[0].values[:-1].reshape(28, 28)

# Wyświetlenie obrazu
plt.imshow(image, cmap='gray')
plt.title('MNIST Data')
plt.axis('off')
plt.show()

# Klastrowanie danych 2D
inertia_2d = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_2d)
    inertia_2d.append(kmeans.inertia_)

# Wizualizacja metody łokciowej dla danych 2D
plt.plot(range(1, 11), inertia_2d, marker='o')
plt.title('Elbow Method for 2D Data')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

# Klastrowanie danych MNIST (nie jest to najlepsza metoda do klastrowania danych MNIST)
inertia_mnist = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_mnist)
    inertia_mnist.append(kmeans.inertia_)

# Wizualizacja metody łokciowej dla danych MNIST
plt.plot(range(1, 11), inertia_mnist, marker='o')
plt.title('Elbow Method for MNIST Data')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Klastrowanie danych 2D
k = 3  # Przykładowa liczba klastrów
kmeans_2d = KMeans(n_clusters=k, random_state=42)
labels_2d = kmeans_2d.fit_predict(data_2d)

# Wizualizacja wyniku grupowania dla danych 2D
plt.scatter(data_2d.iloc[:, 0], data_2d.iloc[:, 1], c=labels_2d, cmap='viridis')
plt.title('Clustering Result for 2D Data')
plt.xlabel('Column 1')
plt.ylabel('Column 2')
plt.colorbar(label='Cluster')
plt.show()

# Redukcja wymiarowości dla danych MNIST za pomocą PCA
pca = PCA(n_components=2)
data_mnist_reduced = pca.fit_transform(data_mnist)

# Klastrowanie danych MNIST po redukcji wymiarowości
kmeans_mnist = KMeans(n_clusters=k, random_state=42)
labels_mnist = kmeans_mnist.fit_predict(data_mnist_reduced)

# Wizualizacja wyniku grupowania dla danych MNIST po redukcji wymiarowości
plt.scatter(data_mnist_reduced[:, 0], data_mnist_reduced[:, 1], c=labels_mnist, cmap='viridis')
plt.title('Clustering Result for MNIST Data (after PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```
